### Linear regression using the Normal Equation with Numpy (without Scikit-Learn estimator)

```python
import numpy as np

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

# 데이터 100개 생성.
# np.random.rand()
# Creates an array of the given shape and populate it with random samples from a uniform distribution over [0, 1) => 0 <= x < 1
X = 2 * np.random.rand(100, 1)
# numpy.random.randn()

# Y = 7 * X + 2의 식을 따라 데이터들이 분포
# Creates an array of specified shape and fills it with random values as per standard normal distribution.
y = 2 + 7 * X + np.random.randn(100, 1)

# X에 저장되어 있는 값
#print(X)

# plot X and y using blue dot markers
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()

# adding a "dummy" input x0 for the bias
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
#print(X_b.shape)
#print(X_b)

# Normal Equation으로 기울기 및 바이어스 구하기
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) # The Normal Equation: (X^T * X)^-1 * X^T * y
print(theta_best)

# new data points
X_new = np.array([[0], [2]])
#print(X_new)
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
#print(X_new_b)
# y = 2 + 7 * X
y_predict = X_new_b.dot(theta_best)
y_predict

plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```

## Performing Linear Regression using Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# np.random.rand()
# Creates an array of the given shape and populate it with random samples from a uniform distribution over [0, 1) => 0 <= x < 1
X = 2 * np.random.rand(100, 1)
# numpy.random.randn()
# Creates an array of specified shape and fills it with random values as per standard normal distribution.
y = 2 + 7 * X + np.random.randn(100, 1)
lin_reg = LinearRegression()
lin_reg.fit(X, y)

#Here
# lin_reg.intercept_ => 'bias'
# lin_reg.coef_ => weights
lin_reg.intercept_, lin_reg.coef_

# plot X and y using blue dot markers
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()

y_predict = lin_reg.predict(X_new)
print(y_predict)

plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```
